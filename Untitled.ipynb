$Create\space several\space graphs,\space implement\space Bellman-Ford\space algorithm,\space check\space its\space complexity$.

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from collections import defaultdict
%matplotlib inline
```

```python
def create_graph(size):
    st = nx.star_graph(size)
    g = nx.DiGraph()
    g.add_nodes_from(st)
    for i in range(1, size + 1):
        g.add_edge(0, i, weight=np.random.randint(-1, 100), directed=True)
    for u,v,w in g.edges(data=True):
        w['weight'] = np.random.randint(-1, 100)
        w['directed'] = True
    node_length = len(g.nodes())
    for i in range(node_length - 1):
        u = np.random.randint(node_length - 1) + 1
        v = np.random.randint(node_length - 1) + 1
        if u != 0 and v != 0:
            g.add_edge(u, v, weight=np.random.randint(-1, 100), directed=True)
    return g
```

```python
def draw_graph(g):
    pos=nx.spring_layout(g)
    nx.draw(g, pos);
    nx.draw_networkx_labels(g, pos);
    edge_labels = dict([((u, v,), d['weight']) for u, v, d in g.edges(data=True)]);
    nx.draw_networkx_edge_labels(g, pos, edge_labels=edge_labels, label_pos=0.2);
```

```python
def check_cycles(g):
    result = 0
    cycles = nx.simple_cycles(g)
    weights = nx.get_edge_attributes(g, 'weight')
    for cycle in cycles:
        cycle.append(cycle[0])
        sumw = sum([weights[(cycle[i-1], cycle[i])] for i in range(1, len(cycle))])
        if sumw < 0:
            result += 1
    return bool(result)
```

```python
def BellmanFord(g, source):
    distances = defaultdict(lambda: 200, {})
    predecessors = defaultdict(lambda: 0, {})
    distances[0] = 0
    nodes = g.nodes()
    edges = g.edges()
    weights = nx.get_edge_attributes(g, 'weight')
    for i in range(len(nodes) - 1):
        for u, v in weights:
            if distances[u] + weights[(u, v)] < distances[v]:
                distances[v] = distances[u] + weights[(u, v)]
                predecessors[v] = u
    return distances, predecessors
```

```python
def create_path(g, predecessors, from_node, to_node=0):
    path = []
    current = from_node
    while current != to_node:
        path.append(current)
        current = predecessors[current]
    path.append(to_node)
    return list(reversed(path))
```

Creating $5$ $graphs$ with different $sizes$.

```python
test_graphs = {}
test_graphs[0] = create_graph(50)
test_graphs[1] = create_graph(100)
test_graphs[2] = create_graph(1000)
test_graphs[3] = create_graph(5000)
test_graphs[4] = create_graph(10000)
```

Checking that there are no $cycles$.

```python
for i in range(5):
    print("Cycles in graph " + str(i) + ": " + str(check_cycles(test_graphs[i])))
```

```python
computing_time = [1.34, 5.16, 554, 17400, 96000]
ev = [2*50**2, 2*100**2, 2*1000**2, 2*5000**2, 2*10000**2]
```

```python
plt.plot(computing_time, [50, 100, 1000, 5000, 10000], color='r', label='Computing time');
plt.plot(np.array(ev)/(np.max(ev)/np.max(computing_time)), [50, 100, 1000, 5000, 10000], color='b', label='C*E*V');
plt.legend(loc='best');
```

As we can see, computing time plot is almost the same as $EV$ plot multiplied by some $constant$, which means that this implementation of Bellman-Ford algorithm is $O(EV)$
